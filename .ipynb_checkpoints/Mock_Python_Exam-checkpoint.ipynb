Question 5:

```python
np.random.seed(123)
list_1 = ["Domestic" for i in range(50)]+["Foreign" for i in range(50)]
list_2 = np.linspace(1,10,100) + np.random.normal(17,2,100)
list_3 = np.round(1000*(np.linspace(10,3,100) + 5.2*np.random.random(100)),2)
autos = pd.DataFrame({"price":list_3,"mpg":list_2,"origin":list_1})
print(autos.head())


avg_by_origin = autos.groupby('origin')['mpg'].mean()
avg_by_origin.plot(kind='bar')
```

Question 6:

```python
# fit the model
xs = autos['mpg']
ys = autos['price']
res = smf.ols('ys ~ xs', data=autos).fit()
res.params

b0 = res.params['Intercept']
b1 = res.params['xs']
print('price = ' + str(b0) +' + ' + str(b1) + 'mpg')
```
